```python
import pandas as pd

players = "Resources/People.csv"
schools = "Resources/Schools.csv"
colleges = "Resources/CollegePlaying.csv"
war2019 = "Resources/mlb2019war.csv"

players_df = pd.read_csv(players)
schools_df = pd.read_csv(schools)
colleges_df = pd.read_csv(colleges)
war2019_df = pd.read_csv(war2019)
```

```python
# new data frame with split value columns 
war2019_df["Player Name"] = war2019_df["Name"].str.split('\\').str[0]

war2019_df["playerID"] = war2019_df["Name"].str.split('\\').str[1]
  
# # making separate first name column from new data frame 
# war2019_df["Player Name"]= new[0] 
  
# # making separate last name column from new data frame 
# war2019_df["Player ID"]= new[1]  
                                   
war2019_df
```

```python
players_df
```

```python
combined_df = pd.merge(war2019_df, players_df, on="playerID")
combined_df
```

```python
colleges_df


college_player_group = colleges_df['playerID'].unique()
# last_college_year = college_player_group['yearID'].max()
last_college_year = []
college_id = []

for player in college_player_group:
    last_college_year.append(colleges_df.loc[colleges_df['playerID'] == player,'yearID'].max())
    
for i in range(0,len(college_player_group)):
    player_specific_df = colleges_df.loc[colleges_df['playerID'] == college_player_group[i],:]
    college_id.append(player_specific_df.loc[player_specific_df['yearID'] == last_college_year[i],'schoolID'].max())

college_player_df = pd.DataFrame(
    {'playerID': college_player_group,
     'Last Year in College': last_college_year,
     'schoolID': college_id,     
    })

college_player_df
```

```python
combined_college_school_df = pd.merge(college_player_df, schools_df, on="schoolID")
combined_college_school_df
```

```python
combined_all_df = pd.merge(combined_df,combined_college_school_df,on="playerID",how="left")
combined_all_df.columns
```
